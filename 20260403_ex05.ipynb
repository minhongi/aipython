```python
student_score = int(input('학생의 점수를 입력하시오.'))

if student_score >= 90:
    print('A')
elif 80 <= student_score < 90:
    print('B')
elif 70 <= student_score < 80:
    print('C')
elif 60 <= student_score < 70:
    print('D')
else:
    print('F')
```

```python
## 실전 사례 분석
import random
answer = random.randint(1, 100)
```

```python
user = int(input('추측값을 입력하시오.'))
```

```python
count=0
while user != answer:
  if user < answer:
    print('더 큰 수를 입력하시오.')
    count+=1
  else:
    print('더 작은 수를 입력하시오.')
    count+=1
  user=int(input('추츳값을 입력하시오.'))
print('총 시도 횟수: ', count)
```
