```python
height1 = float(input('학생의 키(cm)를 입력하시오.'))
height2 = float(input('학생의 키(cm)를 입력하시오.'))

if height1 == height2:
  print('두 학생의 키는 같다.')
if height1 > height2:
  print('첫 번째 학생이 더 크다.')
if height1 < height2:
  print('두 번째 학생이 더 작다.')
```

```python
score = float(input('점수를 입력하시오.'))

if score >= 90:
  print('A')
elif score >= 80:
  print('B')
elif score >= 70:
  print('C')
else:
  print('D')
```

```python
import random

dice1 = random.randint(1,6)
dice2 = random.randint(1,6)

print(f'주사위1:{dice1}')
print(f'주사위2:{dice2}')

if dice1 > dice2:
  print('주사위1 숫자가 더 큽니다.')
elif dice1 == dice2:
  print('주사위1과 주사위2 숫자가 같습니다.')
else:
  print('주사위2 숫자가 더 큽니다.')
```

```python
#가위바위보
import random

man1 = random.randint(1,3)
man2 = random.randint(1,3)

print(f'사람1:{man1}')
print(f'사람2:{man2}')

if man1 == man2:
  print('사람1과 사람2가 비겼습니다.')
elif (man1 == 1 and man2 == 3) or \
     (man1 == 2 and man2 == 1) or \
     (man1 == 3 and man2 == 2):
  print('사람1이 이겼습니다.')
else:
  print('사람2가 이겼습니다.')
```
