## 파이썬 문자열(String) 관련 주요 함수 및 메서드

파이썬의 문자열은 불변(immutable)한 시퀀스 타입으로, 텍스트 데이터를 다룰 때 사용됩니다. 문자열 객체는 다양한 내장 메서드를 제공하여 텍스트를 효율적으로 조작하고 처리할 수 있습니다.

```python
# 예제 문자열
my_string = "  Hello, Python Programming!  "

print(f"원본 문자열: '{my_string}'\n")

# 1. 길이 관련 함수
# len(): 문자열의 길이(문자 수)를 반환합니다.
length = len(my_string)
print(f"1. len(): 문자열 길이 = {length}")

# 2. 대소문자 변환 메서드
# upper(): 모든 문자를 대문자로 변환합니다.
upper_string = my_string.upper()
print(f"2. upper(): 대문자 변환 = '{upper_string}'")

# lower(): 모든 문자를 소문자로 변환합니다.
lower_string = my_string.lower()
print(f"   lower(): 소문자 변환 = '{lower_string}'")

# capitalize(): 첫 글자만 대문자로, 나머지는 소문자로 변환합니다.
capitalize_string = my_string.capitalize()
print(f"   capitalize(): 첫 글자만 대문자 = '{capitalize_string}'")

# title(): 각 단어의 첫 글자만 대문자로 변환합니다.
title_string = my_string.title()
print(f"   title(): 각 단어 첫 글자 대문자 = '{title_string}'\n")

# 3. 공백 제거 메서드
# strip(): 문자열 양쪽 끝의 공백(및 지정된 문자)을 제거합니다.
stripped_string = my_string.strip()
print(f"3. strip(): 양쪽 공백 제거 = '{stripped_string}'")

# lstrip(): 왼쪽 공백 제거
lstripped_string = my_string.lstrip()
print(f"   lstrip(): 왼쪽 공백 제거 = '{lstripped_string}'")

# rstrip(): 오른쪽 공백 제거
rstripped_string = my_string.rstrip()
print(f"   rstrip(): 오른쪽 공백 제거 = '{rstripped_string}'\n")

# 4. 문자열 검색 메서드
# find(sub[, start[, end]]): 부분 문자열이 처음 나타나는 인덱스를 반환합니다. 없으면 -1 반환.
find_index = my_string.find("Python")
print(f"4. find('Python'): 'Python'의 첫 인덱스 = {find_index}")

# rfind(): 부분 문자열이 마지막으로 나타나는 인덱스를 반환합니다.
rfind_index = my_string.rfind("o")
print(f"   rfind('o'): 'o'의 마지막 인덱스 = {rfind_index}")

# count(sub[, start[, end]]): 부분 문자열의 개수를 반환합니다.
count_o = my_string.count("o")
print(f"   count('o'): 'o'의 개수 = {count_o}\n")

# 5. 문자열 교체 메서드
# replace(old, new[, count]): old 문자열을 new 문자열로 교체합니다.
replaced_string = my_string.replace("Python", "Java")
print(f"5. replace('Python', 'Java'): 문자열 교체 = '{replaced_string}'\n")

# 6. 문자열 분할 및 결합 메서드
# split(sep=None, maxsplit=-1): sep을 기준으로 문자열을 분할하여 리스트로 반환합니다.
split_string = stripped_string.split(" ") # 공백 기준으로 분할
print(f"6. split(' '): 공백 기준 분할 = {split_string}")

# join(iterable): iterable의 문자열들을 구분자로 결합하여 새로운 문자열을 반환합니다.
joined_string = "-".join(["apple", "banana", "cherry"])
print(f"   '-'.join(['apple', 'banana', 'cherry']): 문자열 결합 = '{joined_string}'\n")

# 7. 시작/끝 확인 메서드
# startswith(prefix[, start[, end]]): 특정 prefix로 시작하는지 확인합니다.
starts_with_hello = stripped_string.startswith("Hello")
print(f"7. startswith('Hello'): 'Hello'로 시작 = {starts_with_hello}")

# endswith(suffix[, start[, end]]): 특정 suffix로 끝나는지 확인합니다.
ends_with_exclamation = stripped_string.endswith("!")
print(f"   endswith('!'): '!'로 끝남 = {ends_with_exclamation}\n")

# 8. 문자열 구성 확인 메서드
# isalpha(): 모든 문자가 알파벳인지 확인합니다.
print(f"8. 'abc'.isalpha(): {'abc'.isalpha()}")
print(f"   'abc1'.isalpha(): {'abc1'.isalpha()}")

# isdigit(): 모든 문자가 숫자인지 확인합니다.
print(f"   '123'.isdigit(): {'123'.isdigit()}")
print(f"   '123a'.isdigit(): {'123a'.isdigit()}")

# isalnum(): 모든 문자가 알파벳 또는 숫자인지 확인합니다.
print(f"   'abc123'.isalnum(): {'abc123'.isalnum()}")
print(f"   'abc 123'.isalnum(): {'abc 123'.isalnum()}")

# isspace(): 모든 문자가 공백 문자인지 확인합니다.
print(f"   '  '.isspace(): {'  '.isspace()}")
print(f"   ' a '.isspace(): {' a '.isspace()}")

# islower(), isupper(), istitle(): 각각 소문자, 대문자, 타이틀 형식인지 확인합니다.
print(f"   'hello'.islower(): {'hello'.islower()}")
print(f"   'HELLO'.isupper(): {'HELLO'.isupper()}")
print(f"   'Hello World'.istitle(): {'Hello World'.istitle()}\n")

# 9. 서식 지정 (f-string)
name = "Alice"
age = 30
f_string_example = f"제 이름은 {name}이고, 나이는 {age}세입니다."
print(f"9. f-string 예시: {f_string_example}")
```
